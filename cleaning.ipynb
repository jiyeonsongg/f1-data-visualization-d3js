```python
import pandas as pd
import numpy as np
```

# DATA CLEANING

### Q1

```python
q1_df = pd.read_csv('q1_data.csv')
q1_df
```

```python
q1_df_copy = q1_df.copy()
q1_df_copy = q1_df_copy.loc[q1_df_copy['grid'] <= 20].loc[q1_df_copy['grid'] > 0]
q1_df_copy
```

```python
q1_df_copy['position'] = pd.to_numeric(q1_df_copy['position'].replace(r'\\N', 21, regex = True))
# 21 is someone who stop racing
```

```python
q1_df_copy = q1_df_copy.loc[q1_df_copy['position'] <= 21]
```

```python
constructors_df = pd.read_csv('constructors.csv')
drivers_df = pd.read_csv('drivers.csv')
```

```python
merge_q1 = q1_df_copy.merge(constructors_df, how = 'inner', left_on = 'constructorId', right_on= 'constructorId')
merge_q1 = merge_q1.merge(drivers_df, how = 'inner', left_on = 'driverId', right_on= 'driverId')
```

```python
merge_q1
```

```python
q1_output = pd.DataFrame()

q1_output['start'] = merge_q1['grid']
q1_output['finish'] = merge_q1['position']
q1_output['team'] = merge_q1['name']
q1_output['driver'] = merge_q1['surname']
q1_output.to_csv('q1_clean.csv', index=False)
```
