### Q5

```python
qualifying_df = pd.read_csv('qualifying.csv')
qualifying_df
```

```python
qualifying_df = qualifying_df[qualifying_df['driverId'].isin([1, 4, 8])]
```

```python
qualifying_df['driverId'] = qualifying_df['driverId'].replace({1: "Lewis Hamilton", 4: "Fernando Alonso", 8: "Kimi RÃ¤ikkÃ¶nen"})
```

```python
qualifying_df = qualifying_df.replace(r'\\N', np.nan, regex = True).dropna()
```

```python
def convert_mili(time):
    return 60000*int(time[0]) + 100*int(time[2:4]) + int(time[5:8])
```

```python
qualifying_df['q1'] = qualifying_df['q1'].apply(convert_mili)
```

```python
# Hamilton: 2007-2023
# Alonso: 2001-2023
# Raikkonen: 2001-2009 / 2012-2021

# 2008(18 ~ 35) & 2012(860 ~ 879) & 2016(948 ~ 968)
```

```python
qualifying_df_copy = qualifying_df.copy()
qualifying_df_copy['raceId'] = qualifying_df_copy.apply(lambda df: 2008 if df['raceId'] >= 18 and df['raceId'] <= 35 else 2012 if df['raceId'] >= 860 and df['raceId'] <= 879 else 2016 if df['raceId'] >= 948 and df['raceId'] <= 968 else None, axis = 1)
qualifying_df_copy.dropna(inplace=True)
qualifying_df_copy
```
