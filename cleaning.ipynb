# DATA CLEANING

### Q1

```python
q1_df = pd.read_csv('q1_data.csv')
q1_df
```

```python
q1_df_copy = q1_df.copy()
q1_df_copy = q1_df_copy.loc[q1_df_copy['grid'] <= 20]
q1_df_copy
```

```python
q1_df_copy['position'] = pd.to_numeric(q1_df_copy['position'].replace(r'\\N', 21, regex = True))
```

```python
q1_df_copy
```

```python
q1_output = pd.DataFrame()
q1_output['start'] = q1_df_copy['grid']
q1_output['finish'] = q1_df_copy['position']
q1_output.to_csv('q1_clean.csv', index=False)
```
